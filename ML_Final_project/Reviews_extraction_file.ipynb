```python
!pip install google-play-scraper
from google_play_scraper import app, Sort, reviews_all
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Define the list of app IDs you want to scrape
app_ids = [
    'com.blink.crumblecookies',
    'com.insomniacookies.insomnia',
    'com.tossdown.shezan',
    'pk.layers.twa',
    'com.gourmet_foods.GMart',
    'com.blink.pits',
    'com.indolj.hobnob',
    'com.blink.jamilsweets',
    'com.tossdown.jalalsons',
    'com.cuisine.kitchen'
]

# 2. Initialize an empty list to store DataFrames for each app
all_app_dfs = []

# 3. Loop through each app ID and extract reviews
for app_id in app_ids:
    print(f"Scraping reviews for: {app_id}...")
    try:
        # Fetch all reviews for the current app
        reviews = reviews_all(app_id)

        if reviews:
            # Convert list of dictionaries directly into a DataFrame
            df_temp = pd.DataFrame(reviews)

            # Add an extra column to track which app these reviews belong to
            df_temp['app_id'] = app_id

            # Append to our list
            all_app_dfs.append(df_temp)
            print(f"Successfully fetched {len(df_temp)} reviews.")
        else:
            print(f"No reviews found for {app_id}.")

    except Exception as e:
        print(f"An error occurred while scraping {app_id}: {e}")

# 4. Combine all individual DataFrames into one massive DataFrame
if all_app_dfs:
    final_df = pd.concat(all_app_dfs, ignore_index=True)

    # Print summary information
    print("\n--- Scraping Complete ---")
    final_df.info()

    # 5. Export the final consolidated DataFrame to a CSV file
    final_df.to_csv('Multiple_Apps_Reviews.csv', index=False)
    print("Saved all reviews to 'Multiple_Apps_Reviews.csv'")
else:
    print("No reviews were scraped from any of the apps.")
```
